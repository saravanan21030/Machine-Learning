```python
# Import

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# Read the file

data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.head()
```

```python
z = data.iloc[:, [2,3,4]].values
```

```python
z
```

```python
# Dendrogram

dendrogram = sch.dendrogram(sch.linkage(z, method='ward'))
```

```python
# Agglomerative Clustering

model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
model.fit(z)
labels = model.labels_
```

```python
# Scatter plot

plt.scatter(z[labels==0, 0], z[labels==0, 1], s=50, marker='x', color='red')
plt.scatter(z[labels==1, 0], z[labels==1, 1], s=50, marker='x', color='blue')
plt.scatter(z[labels==2, 0], z[labels==2, 1], s=50, marker='x', color='green')
plt.scatter(z[labels==3, 0], z[labels==3, 1], s=50, marker='x', color='purple')
plt.scatter(z[labels==4, 0], z[labels==4, 1], s=50, marker='x', color='orange')

plt.show()
```
